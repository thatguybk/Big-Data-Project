```python
def get_dataframe_info(dataframe, title="Data Frame"):
    dbtable = dataframe
    
    print("Information of the " + title + " Data Frame\n")

    print("First Ten Rows of the Data Frame")
    print(dbtable.head(10))
    
    print("\nSize of Data Frame")
    print(dbtable.shape) #show how much columns and rows in the data frame

    print("\nTypes of Data Stored in the Data Frame\n")
    print(dbtable.dtypes) #show the types of data in the frame

    print("\nGeneral Information on each Column in the Data Frame\n")
    print(dbtable.info()) #give a general view of what the data looks like for each column

    print("\nChecking for Null Rows in the Data Frame\n")
    print(dbtable.isna().all()) #check to see if there is any rows are empty

    print("\nChecking for Null Values in the Columns of the Data Frame\n")
    print(dbtable.isna().any()) #check to see if any value in each row is missing

    print("\nSum of All the Missing Values Per Column in the Data Frame\n")
    print(dbtable.isna().sum()) #return a sum of null number from each column
```

```python
df = pd.read_csv('Raw-Data/Natural_Gas_Import_Quantity.csv')
get_dataframe_info(df, "Natural Gas Import Quantity")
```
